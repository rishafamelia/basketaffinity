# **Basket Affinity per Item**

```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime as dt
```

```python
# upload file

sheet_url = 'https://docs.google.com/spreadsheets/d/106v5vGBPbRT71GFfFXtIJ0TSgRymeGNJkaMy--WjHuU/edit#gid=1844810804'
sheet_url_trf = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
df = pd.read_csv(sheet_url_trf)
df
```

```python
df.info()
```

```python
df2 = df.groupby(['order_id','name'])['quantity'].sum().unstack().reset_index().fillna(0).set_index('order_id')
df2
```

```python
# binary data

def encode_units(x):
  if x <= 0:
    return 0
  if x >= 1:
    return 1

df2_encode = df2.applymap(encode_units)
df2_encode
```

```python
# filter more than 1

df2_plus = df2_encode[(df2_encode > 0).sum(axis = 1) >= 2]
df2_plus
```

```python
pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_df = apriori(df2_plus, min_support=0.03, use_colnames=True).sort_values('support',ascending = False).reset_index(drop = True)

frequent_df['length'] = frequent_df['itemsets'].apply(lambda x: len(x))
frequent_df
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(frequent_df, metric='lift', min_threshold=0).sort_values('lift',ascending = False).reset_index(drop = True)
```

# **Basket Affinity per Category**

```python
df3 = df.groupby(['order_id','category'])['quantity'].sum().unstack().reset_index().fillna(0).set_index('order_id')
df3
```

```python
# binary data

def encode_units(x):
  if x <= 0:
    return 0
  if x >= 1:
    return 1

df3_encode = df3.applymap(encode_units)
df3_encode
```

```python
# filter more than 1

df3_plus = df3_encode[(df3_encode > 0).sum(axis = 1) >= 2]
df3_plus
```

```python
frequent_df3 = apriori(df3_plus, min_support=0.03, use_colnames=True).sort_values('support',ascending = False).reset_index(drop = True)

frequent_df3['length'] = frequent_df3['itemsets'].apply(lambda x: len(x))
frequent_df3
```

```python
association_rules(frequent_df3, metric='lift', min_threshold=0).sort_values('lift',ascending = False).reset_index(drop = True)
```
